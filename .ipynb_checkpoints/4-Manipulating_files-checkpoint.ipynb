# <center>Manipulating files</center>

### <font color=blue>1. [ touch ]</font>
**touch FILE_NAME**

This command will
- create a file with the FILE_NAME with 0 in size, if FILE_NAME does not exist; or
- update the timestamp of the file, if FILE_NAME exists.

### <font color=blue>2. [ cp ]: copy file(s)</font>
**cp [-r] src dst**

Copy a file/files from src to dst. Parameter -r is required when copying directory/directories.

```python
%%bash
### make a copy
# cp bw.inp bw_inp.bak

### copy files somewhere else
# cp bw.inp ~/

### copy files from somewhere else
# cp ~/*.inp .
```

### <font color=blue>3. [ mv ]: move file(s)</font>
**mv src dst**

Move a file/files from src to dst, or rename a file.

```python
%%bash
### rename
# mv bw.inp bw_inp.bak

### move files somewhere else
# mv bw.inp ~/

### move files from somewhere else
# mv ~/*.inp .
```

### <font color=blue>4. [ rm ]: remove file(s)</font>
**rm [-r -f] file1 [file2 file3 \*]**

Remove a file/a directory/files/directories. Parameter -r is required to remove a directory if it is not empty; -f means forcely: deleting without confirm.


<font color=red>** WARNINGS **</font>
- make sure you know what you're going to do with "rm -rf \*"
- double check your rm file list generated by tab-auto-complete function ending with wildcard \*

```python
%%bash
##################### BAD EXP #####################

# my naming system names the working scripts in a way that looks like "zzz-[number]-[description].sh"

# suppose there're 4 files named
### zzz-1-makeInp.sh
### zzz-2-gensel-training.sh
### zzz-3-gensel-validate.sh
### zzz-4-getReport.sh

# suppose I'm going to delete these files

# $ rm z[tab]*    # will work fine

#########################################################################

# things go wrong if there's only 1 file starting with 'z'

# typing rm z[tab]* will eventually became

# $ rm zzz-1-makeInp.sh *

# after hitting the enter key, the nightmare will come if you have not backuped your data
```

### <font color=blue>5. [ ln ]: make link(s) to file(s)</font>
**ln [-s -f] src [dst]**

**ln [-s -f] src1 src2 src3 ... [dst_dir]**

Make a link/links to file/files/directory/directories.
- option -s creates a symbolic link
- option -f means forcely, deleting if dst exists
- deleting link files would not affect the original src files
- deleting the original src files would leave symbol link files broken, normally marked as red by the system in the output of ls


<font color=red>** WARNING **</font>
- Double check your dst parameter if you're linking 2 srcs with -f option

```python
%%bash
##################### BAD EXP #####################

# suppose there're 2 files named
### ../zzz-1-makeInp.sh
### ../zzz-2-gensel-training.sh

# making links 
# $ ln -sf ../z* .  # will work fine

#########################################################################

# things go wrong if I forgot to put a dst '.' for linking 2 files

# typing "ln -sf ../z*" will eventually became

# $ ln -sf ../zzz-1-makeInp.sh ../zzz-2-gensel-training.sh

# after hitting the enter key, the file "../zzz-2-gensel-training.sh" went wrong
```
